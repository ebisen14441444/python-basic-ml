## 教材で使うCSVをダウンロードする

GitHub上の `examples/` フォルダから、4つのCSVを同じ階層にダウンロードします。
Google Colab でも、ローカルでもそのまま動きます。

すでに同名ファイルがある場合は再ダウンロードしません。

```python
from pathlib import Path
from urllib.request import urlretrieve

BASE_URL = "https://raw.githubusercontent.com/ebisen14441444/python-basic-ml/main/examples"
CSV_FILES = [
    "penguins.csv",
    "flights.csv",
    "scores.csv",
    "seattle-weather.csv",
]

data_dir = Path("examples")
data_dir.mkdir(exist_ok=True)

for name in CSV_FILES:
    target = data_dir / name
    if target.exists():
        print(f"skip: {target} (already exists)")
        continue
    url = f"{BASE_URL}/{name}"
    urlretrieve(url, target)
    print(f"downloaded: {target}")
```
